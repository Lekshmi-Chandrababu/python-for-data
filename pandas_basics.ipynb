```python
import pandas as pd

# Creating a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Marks": [85, 92, 78, 88],
    "Attendance": [90, 95, 80, 85]
}

df = pd.DataFrame(data)

print("Student Data:")
print(df)

print("\nBasic Information:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())
```
