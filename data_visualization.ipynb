```python
import matplotlib.pyplot as plt
import pandas as pd

# Creating sample data
data = {
    "Student": ["Alice", "Bob", "Charlie", "David"],
    "Marks": [85, 92, 78, 88]
}

df = pd.DataFrame(data)

# Bar chart
plt.figure()
plt.bar(df["Student"], df["Marks"])
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.title("Student Marks Visualization")
plt.show()
```
