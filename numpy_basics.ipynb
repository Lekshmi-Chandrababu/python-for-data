```python
import numpy as np

# Creating a 1D array
arr = np.array([10, 20, 30, 40, 50])
print("Array:", arr)

# Basic operations
print("Mean:", np.mean(arr))
print("Maximum:", np.max(arr))
print("Minimum:", np.min(arr))

# Creating a 2D array
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("2D Array:\n", matrix)
```
